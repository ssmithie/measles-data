```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('all-measles-rates.csv', index_col=0)
```

```python
df.head()
```

```python
state_df = df.groupby(by='state').mean()
```

```python
state_df
```

```python
state_df.loc[state_df['mmr'] < 0, 'mmr'] = None
```

```python
plt.figure(figsize=(12,10))
sns.scatterplot(x=state_df.index, y='mmr', data=state_df);
```

```python
plt.figure(figsize=(12,10))
sns.boxplot(x=state_df.index, y='mmr', data=state_df, orient='v');
```
